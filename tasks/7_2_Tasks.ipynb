## Задание 1

В качестве первого задания вам предстоит достать значения средних зарплат по городам России. 

Сайт, на котором они хранятся: https://stepik.org/media/attachments/lesson/866758/mean_salary_by_city.html 

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
from tqdm import tqdm
```

```python
response = requests.get("https://stepik.org/media/attachments/lesson/866758/mean_salary_by_city.html", verify=False)
soup = BeautifulSoup(response.content)

data = []
for div in soup.find_all("div", {"class": "row block_rating"}):
    city = (
        div
        .find("div", {"class": "col-xl-7 col-4"})
        .text.strip()
    )
    mean_salary = int(
        div
        .find("div", {"class": "col-xl-3 col-4 text-right"})
        .text.strip()
        .replace('\xa0', '')[:-1]
    )
    data.append((city, mean_salary))
    

res = pd.DataFrame(data, columns=['city', 'mean_salary']).sort_values(by=['city']).reset_index(drop=True)
res.to_csv('results/salary_solution.csv', index=False)
res.head()
```
