**⏲ Напиши свой шедулер!**

```python
def generate_lr_list(start:float, stop:float, iterations:int, step:int)->list():
    generated_lr = []
    num_integer = iterations//step
    num_remainder = bool(iterations%step)
    stop_step = num_integer + int(num_remainder)

    for it in np.linspace(start, stop, stop_step):
         for _ in np.arange(step):
            generated_lr.append(it)

    return generated_lr[:iterations]
```
