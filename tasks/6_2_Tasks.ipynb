# Задача на 5-ом шаге урока

**Разгоняем метамодель**

Ссылка на данные для тестирования

```python
import pandas as pd
import itertools

path = 'https://stepik.org/media/attachments/lesson/825511/meta_feautres_data.csv'
df = pd.read_csv(path)
df.head()
```

```python
rename_columns = {f'{i}': f'target_{i}' for i in df['model'].value_counts().index}

result_df = df.pivot(
    index='index',
    columns='model',
    values='target'
)

models = result_df.columns.tolist()

# Для ВСЕХ упорядоченных пар (включая i=j)
for i in models:
    for j in models:
        result_df[f'{i}_div_{j}'] = result_df[i] / result_df[j]
        result_df[f'{i}_dif_{j}'] = result_df[i] - result_df[j]

result_df.rename(columns=rename_columns).reset_index().to_csv('results/meta_features.csv', index=False)
result_df
```
