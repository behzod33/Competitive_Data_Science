# Задача на 6-ом шаге урока

**Custom eval function в XGBoost**

Решение

```python
# метки получаем методом get_label()
def custom_rmsle_eval(predictions, dmat):
    labels = dmat.get_label()

    a = np.log(predictions+1)
    b = np.log(labels+1)
    loss = np.sqrt(np.mean((a-b)**2))

    return 'RMSLE', loss
```

# Задача на 7-ом шаге урока

**🤯 Extra-hard task: custom Log-cosh loss**

Решение

```python
import pandas as pd
import numpy as np

# метки получаем методом get_label()
def custom_log_cosh_loss(preds, dtrain):
    y = dtrain.get_label()
    n = len(preds)

    delta = preds - y

    grad = n * np.tanh(delta)
    hess = n * (1.0 - np.tanh(delta)**2)

    return grad, hess
```
